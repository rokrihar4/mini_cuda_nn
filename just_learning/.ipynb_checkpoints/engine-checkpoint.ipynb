# Tensors

```python
import torch

x = torch.tensor(2.0, requires_grad=True)
y = x**2 + 3*x

y.backward() # compute gradient

print(x.grad) # dy/dx = 2x + 3

x = torch.tensor([[1, 2], [3, 4]])
print(x.shape)  # (2, 2)
print(x.dim)
```

# Creating a single Neuron

```python
import torch

# A single neuron, literally: 
# y = w * x + b

# Input (2 features)
x = torch.tensor([1.0, 2.0])

# Weights (same size as input)
w = torch.tensor([0.5, -0.5])

# Bias
b = torch.tensor(0.0)

# Neuron output
y = torch.dot(w, x) + b

print(y)

# Adding activation
y = torch.relu(y)
```

```python
# Here we add a activation function ReLU
y = torch.relu(y)
```

```python
# Putting it into GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

x = x.to(device)
w = w.to(device)
b = b.to(device)

y = torch.dot(w, x) + b
y
```

```python
# Make weights learnable
w = torch.tensor([0.5, -0.5], requires_grad=True, device=device)
b = torch.tensor(0.0, requires_grad=True, device=device)

# Target we want
b = torch.tensor(0.0, requires_grad=True, device=device)
target = torch.tensor(1.0, device=device)

# Forward
y = torch.dot(w, x) + b

# Loss (mean squared error)
loss = (y - target) ** 2

# Backprop
loss.backward()

print("Grad w:", w.grad)
print("Grad b:", b.grad)
# This is a tiny neural engine
```

```python
# One training-step
lr = 0.1

with torch.no_grad():
    w -= lr * w.grad
    b -= lr * b.grad

    w.grad.zero_()
    b.grad.zero_()
```
