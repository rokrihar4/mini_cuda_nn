# Creating a single Neuron

```python
import torch

# A single neuron, literally: 
# y = w * x + b

# Input (2 features)
x = torch.tensor([1.0, 2.0])

# Weights (same size as input)
w = torch.tensor([0.5, -0.5])

# Bias
b = torch.tensor(0.0)

# Neuron output
y = torch.dot(w, x) + b

print(y)

# Adding activation
y = torch.relu(y)
```

```python
# Here we add a activation function ReLU
y = torch.relu(y)
```

```python
# Putting it into GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

x = x.to(device)
w = w.to(device)
b = b.to(device)

y = torch.dot(w, x) + b
y
```

```python
# Make weights learnable
w = torch.tensor([0.5, -0.5], requires_grad=True, device=device)
b = torch.tensor(0.0, requires_grad=True, device=device)

# Target we want
b = torch.tensor(0.0, requires_grad=True, device=device)
target = torch.tensor(1.0, device=device)

# Forward
y = torch.dot(w, x) + b

# Loss (mean squared error)
loss = (y - target) ** 2

# Backprop
loss.backward()

print("Grad w:", w.grad)
print("Grad b:", b.grad)
# This is a tiny neural engine
```

```python
# One training-step
lr = 0.1 # learning rate

with torch.no_grad():
    w -= lr * w.grad
    b -= lr * b.grad

    w.grad.zero_()
    b.grad.zero_()
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

# 3 neurons, each takes 2 inputs
W = torch.randn(3, 2, requires_grad=True, device=device)
b = torch.zeros(3, requires_grad=True, device=device)

x = torch.tensor([1.0, 2.0], device=device)   # input (2,)
target = torch.tensor([1.0, 0.0, -1.0], device=device)  # target (3,)

lr = 0.1

for i in range(20):
    # forward pass
    y = W @ x + b   # shape: (3,)

    loss = ((y - target) ** 2).mean()

    # backward
    loss.backward()

    # update
    with torch.no_grad():
        W -= lr * W.grad
        b -= lr * b.grad

    W.grad.zero_()
    b.grad.zero_()

    print(f"{i}. y: {y}, loss: {loss.item()}")
```

```python
# This is meaking a neural network
# You can optionally at a activating function between them
# Input -> [Layer1] -> Activation -> [Layer2] -> Activation -> ... -> Output
device = "cuda" if torch.cuda.is_available() else "cpu"

# Layer 1: 4 neurons, input size 2
W1 = torch.randn(4, 2, requires_grad=True, device=device)
b1 = torch.zeros(4, requires_grad=True, device=device)

# Layer 2: 3 neurons, input size 4 (output of layer 1)
W2 = torch.randn(3, 4, requires_grad=True, device=device)
b2 = torch.zeros(3, requires_grad=True, device=device)

x = torch.tensor([1.0, 2.0], device=device)
target = torch.tensor([1.0, 0.0, -1.0], device=device)

lr = 0.1

for i in range(50):
    # Forward pass
    h = torch.relu(W1 @ x + b1)  # hidden layer + activation
    y = W2 @ h + b2              # output layer (linear for regression)

    # Loss
    loss = ((y - target) ** 2).mean()

    # Backward
    loss.backward()

    # Update weights manually
    with torch.no_grad():
        W1 -= lr * W1.grad
        b1 -= lr * b1.grad
        W2 -= lr * W2.grad
        b2 -= lr * b2.grad

    # Clear gradients
    W1.grad.zero_()
    b1.grad.zero_()
    W2.grad.zero_()
    b2.grad.zero_()

    if i % 5 == 0:
        print(f"{i}. y: {y}, loss: {loss.item()}")
```

```python
# python built-in layers and activations 
import torch
import torch.nn as nn

# Define a simple 2-layer network
class MyNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2, 4)
        self.layer2 = nn.Linear(4, 3)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.layer2(x)
        return x

model = MyNetwork().to(device)

x = torch.tensor([1.0, 2.0], device=device)
target = torch.tensor([1.0, 0.0, -1.0], device=device)

optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
loss_fn = nn.MSELoss()

for i in range(51):
    optimizer.zero_grad()
    y = model(x)
    loss = loss_fn(y, target)
    loss.backward()
    optimizer.step()
    
    if i % 5 == 0:
        print(f"{i}. y: {y}, loss: {loss.item()}")
```

**What is relu?**

ReLU does exactly one thing:
- if the number is negative → replace it with 0
- if the number is positive → leave it alone

That's literally it. In code: max(0, x).

So if your layer outputs:
[-2.1,  0.5, -0.3,  3.2, -1.1]

After ReLU:
[0.0,   0.5,  0.0,  3.2,  0.0]

The reason it exists is without it, stacking layers is pointless. Two linear transforms stacked is just one linear transform. ReLU breaks that linearity so each layer can actually learn something the previous one couldn't.
